```python
# Identify non-date columns
non_date_columns = [
    'RegionID', 'SizeRank', 'RegionName', 'RegionType', 
    'StateName', 'State', 'Metro', 'StateCodeFIPS', 'MunicipalCodeFIPS'
]

# Create a list to hold the transformed records
records = []

# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    # Create a dictionary for the current row
    record = {col: row[col] for col in non_date_columns}
    record['dates'] = {col: row[col] for col in data.columns if col not in non_date_columns}
    
    # Append the record to the list of records
    records.append(record)

cleaned_excel = pd.DataFrame(records)
```

```python
import json
# Identify non-date columns
non_date_columns = [
    'RegionID', 'SizeRank', 'RegionName', 'RegionType', 
    'StateName', 'State', 'Metro', 'StateCodeFIPS', 'MunicipalCodeFIPS'
]

# Create a list to hold the transformed records
records = []

# Iterate over each row in the DataFrame
for _, row in data.iterrows():
    # Create a dictionary for the current row
    record = {col: row[col] for col in non_date_columns}
    record['dates'] = {col.strftime('%Y-%m-%d'): row[col] for col in data.columns if col not in non_date_columns}
    
    # Append the record to the list of records
    records.append(record)

# Convert the list of records to a DataFrame
cleaned_excel = pd.DataFrame(records)

# Convert the DataFrame to a list of dictionaries
records_json = cleaned_excel.to_dict(orient='records')

# Save the list of dictionaries to a JSON file
json_file = 'cleaned_data.json'
with open(json_file, 'w') as f:
    json.dump(records_json, f, indent=4)

print(f'Cleaned data saved to {json_file}')
```
